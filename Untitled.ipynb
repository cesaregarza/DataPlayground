```python
import csv
import pandas as pd
```

```python
all_data = []
with open('arcgis_geocode.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count is 0:
            line_count += 1
            continue
        else:
            line_count += 1
            all_data.append(row)
    print(f'Processed {line_count} lines.')
    df_arcgis = pd.DataFrame(all_data)
```

```python
all_data = []
with open('tamu_geocode.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count is 0:
            headers = row
            line_count += 1
            continue
        else:
            line_count += 1
            all_data.append(row)
    print(f'Processed {line_count} lines.')
    df_tamu = pd.DataFrame(all_data)
    df_tamu.columns = headers
```
