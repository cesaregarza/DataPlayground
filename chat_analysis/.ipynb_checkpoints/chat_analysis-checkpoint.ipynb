```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
f = open("_chat.txt", encoding="utf-8")
input_list = f.read().replace("\u200e","").split("\n")
```

```python
df = pd.DataFrame(input_list)
```

```python
fixed_df = df[0].groupby(df[0].str[0].eq("[").cumsum()).apply(" ".join).to_frame()
regex = r"(^\[\d{1,2}\/\d{1,2}\/\d\d\,\ \d{1,2}\:\d\d\:\d\d\ [A|P]M\])"
fixed_df['date'] = pd.to_datetime(fixed_df[0].str.extract(regex)[0], format="[%m/%d/%y, %I:%M:%S %p]", errors="coerce")
```

```python
bae = fixed_df.loc[1,0][22:25]
author_mask = fixed_df[0].str.contains(bae)
fixed_df['author'] = ""
fixed_df.loc[author_mask, "author"] = bae
fixed_df.loc[~author_mask, "author"] = "Cesar"

fixed_df["lengths"] = fixed_df['date'].dt.strftime("[%m/%d/%y, %I:%M:%S %p]").apply(len) + fixed_df['author'].apply(len) - fixed_df['date'].apply(lambda x: (x.hour < 10) + (x.hour > 12 and x.hour < 22) + (x.day < 10) + (x.month < 10)) + 3
fixed_df.loc[fixed_df['author'] == bae, 'lengths'] += 1
fixed_df['content'] = fixed_df.apply(lambda x: x[0][x['lengths']:],axis=1)

fixed_df = fixed_df.drop(columns=["lengths", 0])
fixed_df = fixed_df.loc[~fixed_df['content'].str.contains("omitted")].reset_index(drop=True)
fixed_df.loc[fixed_df['author'] != "Cesar", "author"] = "Brenda"
```

```python
fixed_df.groupby([pd.Grouper(key="date", freq="D"), pd.Grouper(key="author")]).count().unstack()['content'].plot(ylabel="count")
plt.savefig("figure_1.pdf")
```

```python
hour_df = fixed_df.groupby([fixed_df['date'].dt.hour, pd.Grouper(key="author")]).count().unstack()['content']
```

```python
_, ax = plt.subplots()
hour_df[['Brenda']].plot(kind="bar", ax=ax)
hour_df[['Cesar']].plot(kind="bar", ax=ax, color="coral")
ax.set_xlabel("hour")
ax.set_ylabel("count")
plt.savefig("figure_2.pdf")
```

```python
most_used_words = fixed_df['content'].str.lower().str.split(" ").explode().value_counts()
```

```python
brenda_most_used_words = fixed_df.loc[fixed_df['author'] == "Brenda", "content"].str.lower().str.split(" ").explode().value_counts()
cesar_most_used_words = fixed_df.loc[fixed_df['author'] == "Cesar", "content"].str.lower().str.split(" ").explode().value_counts()
```

```python
cesar_most_used_words = cesar_most_used_words[cesar_most_used_words > 500].reset_index()
brenda_most_used_words = brenda_most_used_words[brenda_most_used_words > 500].reset_index()
```

```python
cesar_most_used_words['unique_ratio'] = cesar_most_used_words.apply(lambda x: x['content'] / most_used_words[x['index']], axis=1)
brenda_most_used_words['unique_ratio'] = brenda_most_used_words.apply(lambda x: x['content'] / most_used_words[x['index']], axis=1)
```

```python
cesar_most_used_words_df = cesar_most_used_words.sort_values("unique_ratio", ascending=False)[:30].reset_index(drop=True)
brenda_most_used_words_df = brenda_most_used_words.sort_values("unique_ratio", ascending=False)[:30].reset_index(drop=True)
```

```python
most_unique_words_df = cesar_most_used_words_df.merge(brenda_most_used_words_df, left_index=True, right_index=True, suffixes=["_cesar", "_brenda"])
most_unique_words_df.columns = ["word_cesar", "count_cesar", "unique_ratio_cesar", "word_brenda", "count_brenda", "unique_ratio_brenda"]
```
