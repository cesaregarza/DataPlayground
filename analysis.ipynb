```python
import pandas as pd
```

```python
#Load data and attempt to drop duplicates right off the bat
app_data_path     = r"google_play_dataset\googleplaystore.csv"
app_df     = pd.read_csv(app_data_path).drop_duplicates().reset_index(drop=True)
```

```python
#Show the values with the most duplicate entries, by the "App" column
apps_by_count = app_df.value_counts("App").loc[lambda x: x > 1]
apps_by_count
```

```python
#Show a view of the dataframe of the app with the most amount of duplicates to investigate how and why it has so many duplicates
app_df.loc[app_df["App"] == apps_by_count.index[0]]
```

```python
#Do the same for the next app
app_df.loc[app_df["App"] == apps_by_count.index[1]]
```

```python
#The differences tend to be in the "Reviews" column for both apps, so we'll make a decision to take the value with the highest number of reviews
review_values = app_df.groupby("App")["Reviews"].max()
review_values
```

```python
#We then filter the dataframe to only include the apps with the highest number of reviews
max_reviews_mask = app_df.groupby("App")["Reviews"].transform(max) == app_df["Reviews"]
app_df.loc[max_reviews_mask]
```

```python
#Make sure the apps are now deduplicated
deduplicated_apps_by_count = app_df.loc[max_reviews_mask] \
                                   .value_counts("App") \
                                   .loc[lambda x: x > 1]
deduplicated_apps_by_count
```

```python
#They aren't, but thankfully it's only a few apps so we can make a case-by-case analysis
duplicate_mask = app_df["App"].isin(deduplicated_apps_by_count.index)
app_df.loc[max_reviews_mask & duplicate_mask].sort_values("App")
```

```python
#We make a list of the row indices that we're going to take. Reasonings:
#Candy Bomb is clearly a game
#Fuzzy Numbers has a genre of Education, so we'll take it
#Learn C++ is clearly educational
#Target has a newer "last updated" column
#Youtube Gaming is entertainment
#osmino has a newer "last updated" column
rows_to_keep = [1754, 742, 670, 2446, 781, 9991]
keep_rows_mask = app_df.index.isin(rows_to_keep)
duplicate_mask = app_df["App"].isin(deduplicated_apps_by_count.index)
duplicate_agg_mask = (~duplicate_mask | keep_rows_mask) & max_reviews_mask
app_df.loc[duplicate_agg_mask]
```

```python
#We check again
app_df.loc[duplicate_agg_mask].value_counts("App").loc[lambda x: x > 1]
```

```python
#We remove apps with invalid ratings and apply the full mask to the dataframe
fixed_df = app_df.loc[duplicate_agg_mask & (app_df["Rating"].notna())]
fixed_df
```

```python
#See if last updated actually has uniform dates
pd.to_datetime(fixed_df["Last Updated"], errors="coerce").loc[lambda x: x.isna()]
```

```python
#Only a single date is broken. Let's check the row
app_df.loc[9990]
```

```python
#Let's check if it has a duplicate value
app_df.loc[app_df["App"] == app_df.loc[9990, "App"]]
```

```python
#It doesn't, so we'll just drop it
fixed_df = fixed_df.drop(9990)
fixed_df["Last Updated"] = pd.to_datetime(fixed_df["Last Updated"])
fixed_df
```

```python
#We then create a new column for the year of the last updated date
fixed_df["Year"]    = fixed_df["Last Updated"].dt.year
fixed_df["Month"]   = fixed_df["Last Updated"].dt.month
fixed_df["Day"]     = fixed_df["Last Updated"].dt.day
fixed_df
```

```python
#Or a dask-friendly version
last_updated_column = pd.to_datetime(fixed_df["Last Updated"])
year_column         = last_updated_column.dt.year .rename("Year")
month_column        = last_updated_column.dt.month.rename("Month")
day_column          = last_updated_column.dt.day  .rename("Day")
pd.concat([fixed_df.iloc[:, :-3], year_column, month_column, day_column], axis=1)
```

```python
#Creating a gross revenue column means fixing the "installs" column, which can be done via str replace
installs_column = fixed_df["Installs"].str.replace(",", "", regex=True) \
                                      .str.replace("+", "", regex=True) \
                                      .str.replace(" ", "", regex=True) \
                                      .str.replace("Free", "0", regex=True) \
                                      .astype(int)
installs_column
```

```python
#No errors, so we now fix the price column
price_column = fixed_df["Price"].str.replace("$", "", regex=True) \
                                .str.replace(",", "", regex=True) \
                                .astype(float)
price_column
```

```python
#Now we get the gross revenue column
gross_revenue_column = price_column * installs_column
gross_revenue_column
```

```python
#And we add it to the dataframe
fixed_df["Gross Revenue"] = gross_revenue_column
fixed_df
```

```python
#Dask-friendly version
pd.concat([fixed_df.drop(columns = ["Gross Revenue"]), gross_revenue_column.rename("Gross Revenue")], axis=1)
```

```python
#Return top 3 gross revenue apps by category
fixed_df_view = fixed_df[["App", "Category", "Gross Revenue"]].set_index("App")
top_by_gross_revenue = fixed_df_view.groupby("Category")["Gross Revenue"].nlargest(3)
top_by_gross_revenue.to_frame()
```

### Thoughts
A number of decisions here were made that can impact the resulting data. The primary one is that the deduplication process involved taking the app with the highest number of reviews. This was a decision, and it's important to take note of the decisions made in any analysis. This might not have been the best choice given infinite time, but it was "good enough" with the given time constraint. Additional decisions were made on a case-by-case basis after the grand sweeping deduplication process went through. It was important to pare down the large amount of duplicates before the decision was taken to use costly human-oversight on the remaining 12 rows.

The general process was aimed more for developer time rather than optimal speed; in particular the reliance on the pd.to_datetime function, while reliable and premade, has a lot of unnecessary overhead. Given more time, deeper investigation of the format of the "Last Updated" column could have led to significantly more performant bespoke code to be written.

All code has been written to be easy to turn to a dask-friendly format, as every single mutable instance has an equivalent thread-safe alternative available within this notebook.
